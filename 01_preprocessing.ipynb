Для начала работы подключается библиотека `pandas`, которая дает возможность работы с данными. Файл с данными был импортирован в среду `Google Colab`. Приступая к работе с библиотекой, выводится первые 20 строк данных с помощью команды `read_csv` и параметра `head()`, выведенные данные имеют разделитель запятую, поэтому параметр `sep` можно не прописывать.

```python
import pandas as pd
df = pd.read_csv('/content/auto.csv')
df.head(20)
```

По данной таблице можно сделать вывод, что предметная область — это поддержанные авто. Представленные в таблице столбцы выглядят следующим образом:

Так в столбце year используется int для указания года машины, в столбцах SellingPrice и kmdriven float64 для указания цены на машину и пройденных километров, а оставшиеся 5 столбцов используют тип данных object.

```python
df.info()
```

Далее используется команда `describe`, которая возвращает основные статистические показатели, состоящие из среднего значения, стандартного отклонения, минимального и максимального значений.

```python
df.describe()
```

Выводятся названия столбцов с помощью `columns`.

```python
df.columns
```

Столбцы необходимо привести к одному виду, каждый должен начинаться с большой буквы, без пробелов или подчеркиваний. Поэтому у столбцов year, kmdriven, fuel, seller_Type, transmission, owner меняются названия с помощью команды `df.rename`.

```python
df = df.rename(columns={
    'year':'Year',
    'sellingprice':'SellingPrice',
    'kmdriven':'KmDriven',
    'fuel':'Fuel',
    'seller_Type':'SellerType',
    'transmission':'Transmission',
    'owner':'Owner'
})

df.columns
```

Делается подсчет кол-ва пропусков в каждом столбце. Видно что пропуски есть в столбцах SellingPrice, KmDriven, Fuel.

```python
df.isna().sum()
```

Решено оставить все строки, где имеются пропуски кроме SellingPrice и заменить пропуски в KmDriven на 0 (у машины может быть нулевой пробег), в Fuel на 'Other' (автомобиль может работать на альтернативном топливе) В конце проведена проверка на пропуски

```python
df = df.dropna(subset=[ 'SellingPrice'])
df['KmDriven'] = df['KmDriven'].fillna(0)
df['Fuel'] = df['Fuel'].fillna('Other')
df.isna().sum()
```

Производится проверка на наличие явных дубликатов и выводится их количество.

```python
print(df.duplicated().sum())
```

Дублирующиеся строки удаляются и проводится проверка правильности удаления.

Чтобы не было пропусков производится обновление индексации, устанавливается аргумент `drop = True`, чтобы не создавался столбец `Index`.

Данные выводятся для проверки.

```python
df = df.drop_duplicates()
print(df.duplicated().sum())

df = df.reset_index(drop=True)

df
```

Теперь проводится проверка данных на наличие неявных дубликатов, для этого используем метод `unique()` для столбцов Fuel, SellerType, Transmission, Owner, которые имеют категориальные переменные.

```python
print(df['Fuel'].unique())
print(df['SellerType'].unique())
print(df['Transmission'].unique())
print(df['Owner'].unique())
```

Можно наблюдать повторяющиеся элементы в двух столбцах из четырех. Это исправляется путем задания им одинаковых имен. Код сразу проверяется на правильность.

```python
df['Fuel'] = df['Fuel'].replace('Petroll','Petrol')
df['SellerType'] = df['SellerType'].replace('Dealeer','Dealer')

print(df['Fuel'].unique())
print(df['SellerType'].unique())
```

Вывод информации о датасете.

```python
df.info()
```

Далее производится группировка и создание сводных таблиц в соответствии с вариантом №11.

**Задание 1:** Группировка - тип владельца (owner) по количеству разных коробок передач (transmission)

```python
df.groupby(['Owner', 'Transmission']).size()
```

По результатам группировки можно сделать вывод, что на вторичном рынке автомобилей преобладают машины с механической коробкой передач, что косвенно может нам сказать, что люди не могут позволить себе машины на автомате, так как комплектации с этой коробкой передач всегда отличаются в цене в высшую сторону.

Также можно увидеть, что машин с одним или двумя владельцами значительно больше остальных.

**Задание 2:** Группировка - тип владельца (owner) по количеству разных коробок передач (transmission).
Создать датафрейм. Переименовать столбец с количеством в “сount”. Отсортировать по убыванию
столбца “count”.

```python
df_grouped = df.groupby(['Owner', 'Transmission']).size().reset_index(name='count')
df_sorted = df_grouped.sort_values(by='count', ascending=False)
display(df_sorted)
```

Эта группировка ничего кардинально не меняет, просто сортировка позволяет более наглядно увидеть, что на рынке доминируют автомобили с механической коробкой передач.

**Задание 3:** Сводная таблица (pivot_table) - среднее количество пройденных километров для каждого
типа года year. Отсортировать по убыванию. Округлить до двух знаков.


```python
df.pivot_table(index=['Year'],
               values='KmDriven',
               aggfunc='mean').sort_values(by='KmDriven', ascending=False)['KmDriven'].round(2)
```

По таблице можно сделать вывод, что чем старше автомобиль, тем больше у него средний пробег, однако внизу таблицы мы можем увидеть 1996 или 1999 год, у этих машин средний пробег меньше чем например у машин 2013 или 2014 года. Довольно занятный случай.

**Задание 4:** Сводная таблица (pivot_table) - среднее количество пройденных километров для каждого года - строки и типа владельца - owner - столбцы. Отсортировать по возрастанию year. Округлить до
двух знаков.


```python
df.pivot_table(index=['Year'],
               columns='Owner',
               values='KmDriven',
               aggfunc='mean').sort_values(by='Year', ascending=True).round(2)
```

Данная сводная таблица позволяет посмотреть средний пробег машин каждого года, еще и со знанием количества владельца, где `NaN` значит в такой год машин с пробегом нет. В основном чем больше количество владельцев, тем больше пробег у машины, но бывают и исключения. Также можно понять, что машины выдающиеся на тест драйв не старше 2017 года.

### Задание 1

Сделать категоризацию столбца price, добавить минимум 3 категории.

```python
price_labels = ['Low', 'Medium', 'High', 'SuperHigh']
df['PriceCategory'] = pd.cut(df['SellingPrice'], bins= [0, 200000, 500000, 1000000, df['SellingPrice'].max()], labels=price_labels)

print(df[['SellingPrice', 'PriceCategory']].head(25))
```

Был создан новый столбец `PriceCategory`, в котором реализована категоризация цен на автомобили `SellingPrice`. Всего создано 4 категории, где `Low` - от 0 до 200000, `Medium` - от 200000 до 500000, `High` - от 500000 до 1000000, `SuperHigh` - цены от 1000000 и выше.

### Задание 2

По каждой категории из пункта 1 продемонстрировать средний, минимальный, максимальный пробег

```python
mileage_stats = df.groupby('PriceCategory', observed=True)['KmDriven'].agg(['mean', 'min', 'max'])

print(mileage_stats)
```

Чем выше класс автомобиля, его цена, тем ниже его средний пробег. Показатели минимального и максимального пробега не так однозначны, здесь не прослеживаются какие-либо взаимосвязи. Так, можно подметить, что максимальный и минимальный пробег имеют автомобили из категории `Medium`.
